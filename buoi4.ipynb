```python
count = 0
f = open("input.txt", "r", encoding="utf-8")
content = f.read()
for word in content:
    if word == "\n":
        count += 1
print(count)
f.close()
```

```python
with open("input.txt", "r", encoding="utf-8") as f:
    content = f.read()

copy = open("copy.txt", "x", encoding="utf-8")
copy.write(content)
copy.close()
```

```python
with open("copy.txt", "r", encoding="utf-8") as f_in:
    lines = []
    for i in range(5):
        line = f_in.readline()
        lines.append(line)

with open("output.txt", "w", encoding="utf-8") as f_out:
    f_out.writelines(lines)
    if lines:
        if lines[-1][-1] != "\n":
            f_out.write("\n")
    f_out.write("Họ và tên\n")
```
